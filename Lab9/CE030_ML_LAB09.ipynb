## Implement K Means Clustering Algorithm on the given dataset

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples = 100,centers=3,n_features=2,cluster_std=0.2,random_state=0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0],X[:,1], c='white',marker='o',edgecolor='black',s=50)
plt.show()
```

###using scikit-learn to perform K-means clustering

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3, init='random',n_init=10,max_iter=300, tol=1e-04, random_state=0)
y_km = km.fit_predict(X)
```

```python
plt.scatter(
    X[y_km == 0,0],X[y_km == 0,1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[y_km == 1,0],X[y_km == 1,1],
    s=50, c='orange',
    marker='o',edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[y_km == 2,0],X[y_km == 2,1],
    s=50,c='lightblue',
    marker='v',edgecolor='black',
    label='cluster 3'
)
```


###plot the centroids

```python
plt.scatter(
    km.cluster_centers_[:,0],km.cluster_centers_[:,1],
    s=250, marker='*',
    c='red',edgecolor='black',
    label='centroids'
)

plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.metrics import silhouette_score
```

### silhouette_score

```python
score = silhouette_score(X,km.labels_,metric='euclidean')
print('Silhouetter Score: %.3f' % score)
```

## Hand Written Digit Recognition

```python
import numpy as np 
import pandas as pd 
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mode
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import scale
import time
```

```python
digits = load_digits() 
```

```python
digits.data.shape
```

```python
digits.keys()
print(digits.data.size)
print(digits.keys())
print(digits.data)
print(digits.target)
print(digits.target.size)
```

```python
digits.data[0:3]
```

```python
km = KMeans(n_clusters=10,random_state=0)
X = digits.data
Y = digits.target
y_km = km.fit_predict(X)
```

```python
kmeans_cluster = KMeans(init="random", n_clusters=10, n_init=10, random_state=0)
kmeans_cluster.fit(X)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
reduced_data = pca.fit_transform(X)
kmeans_cluster.fit(reduced_data)
```

```python
centroids = kmeans_cluster.cluster_centers_
label = kmeans_cluster.fit_predict(reduced_data)
unique_labels = np.unique(label)
```

```python
plt.figure(figsize=(8, 8))
for i in unique_labels:
    plt.scatter(reduced_data[label == i, 0],
                reduced_data[label == i, 1],
                label=i)
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='k', zorder=10)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2, 5, figsize = (8,3))
centers = km.cluster_centers_.reshape(10,8,8)
for axi, center in zip(ax.flat, centers):
  axi.set(xticks = [], yticks = [])
  axi.imshow(center, interpolation='nearest',cmap = plt.cm.binary)
```
